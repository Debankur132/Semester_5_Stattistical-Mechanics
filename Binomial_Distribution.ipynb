```python
import numpy as np
import matplotlib.pyplot as plt
def C(N,x):
  return np.math.factorial(N)/(np.math.factorial(x)*np.math.factorial(N-x))
def P(N,x,p):
  return C(N,x)*p**x*(1-p)**(N-x)
```

```python
p=0.5 #as stated in the problem
```

```python
N=20 
x1=np.arange(N+1)
P1=np.zeros(len(x1))
for i in range(len(P1)):
  P1[i]=P(N,x1[i],p)
```

```python
N=200
x2=np.arange(N+1)
P2=np.zeros(len(x2))
for i in range(len(P2)):
  P2[i]=P(N,x2[i],p)
```

```python
#PLotting
plt.subplot(2,2,1)
plt.plot(x1,P1,'r')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.title("For N = %d"%20)

plt.subplot(2,2,2)
plt.plot(x2,P2,'g')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.title("For N = %d"%200)

plt.subplot(2,2,3)
plt.plot(x1,np.log(P1),'b')
plt.xlabel('x')
plt.ylabel('ln P(x)')

plt.subplot(2,2,4)
plt.plot(x2,np.log(P2),'c')
plt.xlabel('x')
plt.ylabel('ln P(x)')
plt.show()
```
